```python
class Solution:
    def sampleStats(self, count: List[int]) -> List[float]:
        # Good: learn how to code
        minimum = float("inf")
        maximum = 0
        total_sum = 0
        total_count = sum(count)
        mode_value = 0
        mode_count = 0
        mean = 0
        median = 0

        median_count = total_count // 2
        median_found = False

        for value, freq in enumerate(count):
            if freq > 0:
                minimum = min(minimum, value)
                maximum = max(maximum, value)
                total_sum += value * freq

                if freq > mode_count:
                    mode_value = value
                    mode_count = freq

                mean += value * freq

                if not median_found and total_count % 2 == 0:
                    if median_count == 0:
                        median += value
                    elif median_count < freq:
                        median += value / 2
                        median_found = True
                    else:
                        median_count -= freq
                elif not median_found and total_count % 2 != 0 and median_count < freq:
                    median = value
                    median_found = True

        mean /= total_count

        return [minimum, maximum, mean, median, mode_value]
```
